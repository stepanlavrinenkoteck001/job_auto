```python
def print_line_breaks(str, n=80):
    [print(str[i:i+n]) for i in range(0, len(str), n)]
```

```python

def compare_gpts_printout(df, job_post_index, n=80):
    print_line_breaks(df['description'][job_post_index], n = n)
    print('###############')
    print_line_breaks(df['hard_skillsggml-gpt4all-j-v1.3-groovy'][job_post_index], n=80)
    print('##############')
    print_line_breaks(df['hard_skillschatgpt'][job_post_index], n = n)
```
