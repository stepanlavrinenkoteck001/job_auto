```python
gpt_model = gpt4all.GPT4All(gpt_model_str, model_path=model_path)
```

```python
spacy_pipeline = spacy_cleaner.Pipeline(
    model,
    removers.remove_stopword_token,
    removers.remove_punctuation_token,
)

```

```python

def gpt4all_prompt(gpt_model, question_string, description, spacy_prep = False):
    if spacy_prep:
        description = spacy_pipeline.clean([description])[0]
    prompt_string = description + '. '+ question_string

    messages = [{"role": "user", "content": prompt_string}]

    ret = gpt_model.chat_completion(default_prompt_footer = False, 
                            default_prompt_header = False,
                            messages = messages,
                            verbose = False)
    if ret["choices"][0]["message"]['content'] == '':
        Warning('Empty string')
        return np.nan
    else:
        return ret["choices"][0]["message"]['content']

        

```

```python
def gpt4all_prompt_small(gpt_model, message):

    ret = gpt_model.chat_completion(default_prompt_footer = False, 
                            default_prompt_header = False,
                            messages = message,
                            verbose = False,
                            streaming = False)
    if ret["choices"][0]["message"]['content'] == '':
        Warning('Empty string')
        return np.nan
    else:
        return ret["choices"][0]["message"]['content']
```

```python
def prompt_str(description, question_string):   
    return description + '. '+ question_string
```

```python
messages = [{"role": "user", "content": prompt_str(description, question_string)} for description in df['description']]
```

```python
returns = [gpt4all_prompt_small(gpt_model, [message]) for message in messages]
```
